```python
# 라이브러리 호출
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import chromedriver_autoinstaller
import requests # 호출 위함
import sys
import os
import re
import pandas as pd
import numpy as np
```

```python
# 현대자동차 전체 모델 리스트

car_codes = pd.read_csv('차종-모델 코드.csv',encoding='utf-8')
car_codes
```

```python
# best trim 없는 차량 삭제

car_codes = car_codes[car_codes.modelCode!='없음']
car_codes
```

```python
model = car_codes['model'].values.tolist()
model_code = car_codes['modelCode'].values.tolist()
print(model)
print(model_code)
print(len(model))
print(len(model_code))
```

```python
# chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(r"C:\Users\chromedriver.exe")

color_information=[]
for m1,m2 in zip(model,model_code):
    url = 'https://www.hyundai.com/kr/ko/e/vehicles/estimation/model/making?model='+m1+'&modelCode='+m2
    driver.get(url)
    response = requests.get(url)
    html = response.text
    
    color_list = []
    x = driver.find_elements(By.CSS_SELECTOR,'#pane-recomTab0 > div > ul > li')
    name = driver.find_element(By.CSS_SELECTOR,'#making > div.header-wrap > header > div.logo-wrap > button > span').text
    model = driver.find_element(By.CSS_SELECTOR,'#making > div.content-wrap > div > div > section.preview > div.preview-model > header > div.heading > span').text
    color_list.append(name)
    color_list.append(model)
    for i in range(1,len(x)+1):
        colors = driver.find_element(By.CSS_SELECTOR, '#pane-recomTab0 > div > ul > li:nth-child('+str(i)+') > div > div > span').text
        colors_text = driver.find_element(By.CSS_SELECTOR,'#pane-recomTab0 > div > ul > li:nth-child('+str(i)+') > p').text
        color_list.append(colors_text+' '+colors)
    color_information.append(color_list)
        
color_information
```

```python
driver.quit()
```

```python
for color_info in color_information:
    if len(color_info)==3:
        color_info.append('-')
        color_info.append('-')
    elif len(color_info)==4:
        color_info.append('-')
    else:
        pass
    
color_information
```

```python
columns = ['name','model','color_1st','color_2nd','color_3rd']
df_color = pd.DataFrame(color_information,columns=columns)
df_color
```

```python
df_color.to_csv('데이터 수집.csv',encoding='cp949') # 데이터 수집 완료
```
